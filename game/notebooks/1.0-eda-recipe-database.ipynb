# 1.0 Investigate recipe table

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/recipe_table.csv', sep=';')
df.info()
```

```python
df.head(2)
```

```python
plt.figure(figsize=(20,20))
plt.subplot(221)
sns.distplot(df['calories'])
plt.title('Calories')

plt.subplot(222)
sns.distplot(df['carbs'])
plt.title('Carbs')

plt.subplot(223)
df['cuisine'].value_counts().plot(kind='bar')
plt.title('Cuisine')

plt.subplot(224)
df['food_group'].value_counts().plot(kind='bar')
plt.title('Food group')
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
df['season'].value_counts().plot(kind='bar')
plt.title('Season')

plt.subplot(122)
sns.distplot(df['cooking_time'], kde=False)
plt.title('Cooking time')
```

```python
plt.figure(figsize=(20,20))
plt.subplot(221)
sns.distplot(df['price_1p_pence'], kde=False)
plt.title('Price 1 person pence')

plt.subplot(222)
sns.distplot(df['price_2p_pence'], kde=False)
plt.title('Price 2 person pence')

plt.subplot(223)
sns.distplot(df['price_4p_pence'], kde=False)
plt.title('Price 4 person pence')
```
